```python
# Import Statements
import pyrealsense2 as rs
import numpy as np
import cv2

# Configure depth and color streams
pipeline = rs.pipeline()
config = rs.config()
config.enable_stream(rs.stream.gyro)
config.enable_stream(rs.stream.accel)

profile = pipeline.start(config)

try:
    while True:
        frames = pipeline.wait_for_frames()
        
        accel_sample = frames[0].as_motion_frame().get_motion_data()
        print("Accel: ", accel_sample.x, " ", accel_sample.y, " ", accel_sample.z, "\n")
        gyro_sample = frames[1].as_motion_frame().get_motion_data()
        print("Gyro: ", gyro_sample.x, " ", gyro_sample.y, " ", gyro_sample.z, "\n")
        
        key = cv2.waitKey(1)
        if key & 0xFF == ord('q') or key == 27:
            cv2.destroyAllWindows()
            break
        
finally:

    # Stop streaming
    pipeline.stop()
```
